### Frequency Distribution
- 단어의 빈도수를 활용할 수 있는 많은 메소드를 제공
- 인간의 언어 -> 컴퓨터가 인식
    - 전체 단어 수
    - 특정의 빈도수

```python
fdist = FreqDist(text1) # nltk.FreqDist()
fdist # 딕셔너리 = key:value
```

```python
fdist.most_common(10) # 가장 빈번한 단어 10개만 
```

```python
fdist['whale'], text1.count('whale')
```

```python
fdist.freq('whale') #전체 텍스트 중 특정 단어의 비율
```

```python
fdist.plot(50, cumulative=True) # 가장 빈번한 것
```

```python
fdist.plot(50, cumulative=False)
```

### list comprehension
- 지능형 리스트
- 주어진 리스트를 포함하는 새로운 리스트를 형성하는 것
- list + for_loop 한 줄로 해결하는 방식
- [A for A' in B]
    - B에 있는 A'에서 A를 가진 리스트를 반환
    - B: 기존의 리스트
    - A': B의 자료 형태
    - A: 출력값, 새로운 리스트의 요소
    
- 조건+ : [A for A' in B if <조건식>]
    - [A for A' in B if<조건식> and ]

```python
[d for d in food]
```

```python
[d[0] for d in food]
```

```python
# text1 알파벳 수가 10이상인 단어
[word for word in text1 if len(word) >= 10]
```

```python
# text4에서 'gnt'를 포함하는 단어 반환
[word for word in text4 if word.find('gnt') > 0] #
[word for word in text4 if 'gnt' in word] #
```

```python
word1 = [word for word in set(text4) if 'gnt' in word] # 중복을 없애고 찾음 > list
    # the, The 공존 -> 연산 | 중복의 기준이 먼저 들어감
word2 = set([word for word in text4 if 'gnt' in word]) # 찾고 중복을 없앰 > set
    # 연산 -> 중복 없앰

len(word1), len(word2)
```

### 리스트. 스트링

```python
sent = 'my, name, is, 00'
# ['my', 'name', 'is', '00'] tokens

sent.strip().split(',') # 공백 제거가 안됨.
# sent.split(',').strip() 은 안 됨 str.strip()만 가능

tokens = sent.split(',')
[word.strip() for word in tokens]
```
